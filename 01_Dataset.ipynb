```python
def plot_serie_temporal(ax,
                        x,
                        y,
                        legenda = '',
                        xlabel = '',
                        ylabel = '',
                        linewidth = 2,
                        linestyle = '-',
                        marker = ''):
    '''
    Plota uma série temporal (x,y) em ax.
    '''
    ax.plot(x, y, linewidth=linewidth, linestyle=linestyle, label=legenda, marker=marker)
    ax.grid(True)

    if(len(xlabel) > 0):
        ax.set_xlabel(xlabel)
    if(len(ylabel) > 0):
        ax.set_ylabel(ylabel)
    if(len(legenda) > 0):
        ax.legend(loc='best')

    return ax
```

```python
#Criação de uma função para deixar os dataframes relacionados as propriedades retiradas
#da estação automática de Itaporanga do mesmo tamanho dos dataframes relacionados as
#propriedades retiradas da estação automática de Patos

def comp_it (nomecoluna,
             cidadecoluna):
    '''
    Completar com NaN os dados ausentes de Itaporanga para se enquadrar no tamanho do DF de Patos
    '''
    
    #Criação de dados NaN para completar as datas entre o início de operação 
    #da estação de Patos até o início de operação da estação de Itaporanga
    To_Complete_Before = pd.DataFrame(index = range(0, 90936), columns = nomecoluna)
    
    #Criação de dados NaN para completar completar até a data da última medição
    #já que a última medição de Itaporanga foi bem antes (em meados de Abril) da de Patos
    To_Complete_After = pd.DataFrame(index = range(0, 1954), columns = nomecoluna)
    
    #Transformando a coluna de um dataframe em um dataframe
    cidadecoluna = pd.DataFrame(cidadecoluna)
    
    #Pré-processo para a concatenação de dataframes
    Completar = [To_Complete_Before , cidadecoluna, To_Complete_After]
    
    #Concatenação do dataframe
    Completo = pd.concat(Completar, ignore_index = 'False')
        
    return Completo
   
```

```python
#Criação de uma função pra deixar um dataframe único com as propriedades obtidas de cada
#estação automática e sua média

def df_prop(df,                    #Nome do DATAFRAME a ser criado
           nomecolunaoriginal,     #Nome da coluna da propriedade de acordo com o INMET
           prop_sg,                #Nome do DATAFRAME já criado para a propriedade em SÃO GONÇALO
           prop_ita):              #Nome do DATAFRAME já criado para a propriedade em ITAPORANGA
    
    '''
    Adicionando colunas em um dataframe
    '''
    
    #Criação de uma lista vazia
    df = []
    
    #Transformação dessa lista vazia em um dataframe vazio
    df = pd.DataFrame(df)
    
    #Adição das colunas no dataframe
    df['DATA'] = Unique['DATA']
    df['PATOS'] = Patos[nomecolunaoriginal]
    df['SAO GONCALO'] = prop_sg[nomecolunaoriginal]
    df['ITAPORANGA'] = prop_ita[nomecolunaoriginal]

    #Construção de uma coluna que tenha um média da radiação de cada estação para cada medição
    df['MEDIA'] = pd.DataFrame(df.mean(axis=1))

    propriedade = pd.DataFrame(df)
    
    return propriedade
```

```python
#Definição de uma função que plote a série temporal de alguma propriedade, e de sua média

def plot_media(x,                #Componente do eixo x (único para todas as propriedades)
               yP,               #Propriedade de PATOS
               ySG,              #Propriedade de SÃO GONÇALO
               yI,               #Propriedade de ITAPORANGA
               yM,               #Propriedade MÉDIA
               ylabel,           #Título do eixo y
               title):           #Título da imagem
    
    '''
    Plotar as 3 componentes para cada cidade e a média.
    '''
    
    #Definição da quantidade de "subimagens" dentro da
    #imagem, tamanho da imagem e da cor do background.
    fig, axs = plt.subplots( 2, figsize = (15, 8), facecolor = "w" )

    #Definição do título.
    fig.suptitle(title)

    #O que será plotado na primeira subimagem, no caso, a componente para cada cidade,
    #sendo que as cidades já estarão pré-definidas, então quando utilizar a função
    #utilizar na ordem PATOS, SÃO GONÇALO E ITAPORANGA.
    axs[0].plot(x, yP, 'tab:blue',label = 'Patos')
    axs[0].plot(x, ySG, 'tab:orange',label = 'São Gonçalo')
    axs[0].plot(x, yI, 'tab:green',label = 'Itaporanga')

    #Definição do título do eixo Y da primeira subimagem
    axs[0].set_ylabel(ylabel)

    #Definição de onde ficará localizada a legenda na primeira subimagem
    axs[0].legend(loc = 'best')

    #O que será plotado na segunda subimagem, no caso, a média de todas as cidades.    
    axs[1].plot(x, yM, 'tab:cyan' ,label = 'Média', linestyle = '--')

    #Definição do título do eixo Y da segunda subimagem
    axs[1].set_ylabel(ylabel)
    
    #Definição de onde ficará localizada a legenda na segunda subimagem  
    axs[1].legend(loc = 'best')


    return fig.show()
```

```python
#Criação de um DataFrame contendo a média diária da média da propriedade GHI (W/m²)
#para os dados do NREL 

#Criação de um dicionário vazio
GHI_SUM_Dia = {}

#Transformação de um dicionário vazio em um DataFrame vazio
GHI_SUM_Dia = pd.DataFrame(GHI_SUM_Dia)

#Transformação de uma coluna de um DataFrame em um DataFrame propriamente dito
GHI_Media = pd.DataFrame(Coremas_NSRDB['GHI'])

#Estrutura de repetição para que a média seja feita para cada 24 horas, com uma iteração
#de 48 slots devido o frequência de amostragem ser de 30 minutos
for i in range(0, int(len(Coremas_NSRDB['GHI'])/48)):
    GHI_SUM_Dia[i] = GHI_Media.iloc[(i * 48): ((i + 1)*48)].sum(axis = 0)

#O DataFrame GHI_SUM_Dia[i] nos retornará um dataframe em que os dados são dispojados em linha
#então, fazendo a transposta, teremos um dataframe com os dados dispojados em coluna

#Esse novo dataframe será a média diária, transformando W/m² em kWh/m²dia
GHI_MED_Dia = GHI_SUM_Dia.transpose()

#Dividindo todos os dados do dataframe por 24 (horas por dia) * 1000 (para transformar em k)
GHI_MED_Dia = GHI_MED_Dia.div(24000) 

#Multiplicando todos os dados do dataframe for 12 (número de horas de sol em um dia)
GHI_MED_Dia = GHI_MED_Dia.mul(12)

#Renomeação da coluna
GHI_MED_Dia = GHI_MED_Dia.rename(columns = {'GHI': 'MEDIA DIARIA - NREL (kWh/m²/dia)'})

GHI_MED_Dia

#Importação das datas do dados
Data_NREL = pd.read_excel('/Coremas_NSRDB_Data.xlsx')

#Criação de um dicionário vazio
GHI_NREL = {}

#Transformação de um dicionário vazio em um DataFrame vazio
GHI_NREL = pd.DataFrame(GHI_NREL)

#Inserção da coluna data referentes as datas de medições do NREL
GHI_NREL['DATA']= Data_NREL['DATA']

#Inserção da coluna da média diária da GHI do NREL
GHI_NREL['MEDIA DIARIA - NREL (kWh/m²/dia)'] = GHI_MED_Dia['MEDIA DIARIA - NREL (kWh/m²/dia)']

GHI_NREL
```
